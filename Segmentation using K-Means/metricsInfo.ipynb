```python
import numpy as np
import pandas as pd
```

```python
path_to_metrics = '/Users/anastasia/Desktop/nnn/metrics.txt'

file1 = open(path_to_metrics, 'r')
Lines = file1.readlines()

content = []
for line in Lines:
    content.append(line.strip().split())
```

```python
df = pd.DataFrame(content, columns =['name', 'accuracy', 'optimal_k'])
df['optimal_k'] = pd.to_numeric(df['optimal_k'])
df['accuracy'] = pd.to_numeric(df['accuracy'])
df
```

```python
df.sort_values(['optimal_k'], inplace=False)
df['optimal_k'].hist()
```

```python
df.sort_values(['accuracy'], inplace=False)
print(f"mean = {df['accuracy'].mean()}")
df['accuracy'].hist()
```

```python
df[df['accuracy']<=0.4].hist()
```
